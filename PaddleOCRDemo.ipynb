```python
import cv2
from rapidocr_onnxruntime import RapidOCR

rapid_ocr = RapidOCR()

image_path = '2.jpg'
img = cv2.imread(image_path)
ocr_result, _ = rapid_ocr(img)

for v in ocr_result:
    print(v)
```
